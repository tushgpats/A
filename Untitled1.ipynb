```python
import requests
import pandas as pd
movie_file = pd.read_json("movies.json","r")
for movies in movie_file["data"]:
    url = 'http://www.omdbapi.com/?t='+movies["name"]+'&y='+str(movies["year"])+'&apikey=151ef65e'
    r = requests.get(url)
    print(r.text)
    json_data = r.json()
    dadata = {}
    for k in json_data.keys():
        print(k + ':', json_data[k])
        file = open("movie_details","a",encoding = "utf-8")
        file.write(str(dadata))
        file.close()

       
#http://www.omdbapi.com/?i=tt3896198&apikey=151ef65e
```
